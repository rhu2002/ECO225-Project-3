### Crime Metric Definitions
**Theft Frequency:** The number of theft occurences within a district

**Theft Rate:** The percentage of crimes that were thefts within a district

**Theft Arrest Rate:** The percentage of theft crimes that resulted in arrest within a district

**Crime Arrest Rate:** The percentage of crimes that resulted in arrest within a district

**Share of Theft:** The percentage of all thefts that occured within a district

**Thefts Per Capita:** The number of thefts per person within a district

**Crimes Per Capita:** The number of crimes per person within a district

**Gun Offender Frequency:** The number of registered gun offenders within a district

**Gun Offenders Per Capita:** The number of registered gun offenders per person within a district

```python
all_district_data = [c_list, e_list, ne_list, n_list, nw_list, se_list,
                     s_list, sw_list, w_list] # Create a list of all district data

# Nested for loop to access each population value to convert to type integer
for district in all_district_data:
  for idx, value in enumerate(district):
    new_string = value.replace(",", "")
    final_value = int(new_string)
    district[idx] = final_value
```

```python
income_levels = [10000, 15000, 25000, 35000, 45000, 55000, 68000, 88000,
                 113000, 138000, 175000, 200000] # Create income levels by midpoint of each income bracket
total_population_by_district = [] # Create total population by district list

# For loop through to add each district's total population to the list 
for district in all_district_data:
    total_population_by_district.append(sum(district))
    
avg_income_by_district = [] # Create average income by district list

# For loop through to calculate the average income by district and append to the list
for i, district_values in enumerate(all_district_data):
  num = sum(np.multiply(district_values, income_levels)) / total_population_by_district[i]
  avg_income_by_district.append(num)
```

```python
district_names = ['CENTRAL', 'EASTERN', 'NORTHEASTERN', 'NORTHERN',
                  'NORTHWESTERN', 'SOUTHEASTERN', 'SOUTHERN',
                  'SOUTHWESTERN', 'WESTERN'] # Create a list of district names
avg_income_by_district = pd.Series(avg_income_by_district,
                                   index=district_names,
                                   name='Average Annual Household Income (USD)') # Convert the list into a series
populations = pd.Series(total_population_by_district, index=district_names,
                        name='Total Population') # Convert the list into a series
income_and_pop_df = pd.merge(avg_income_by_district, populations,
                             right_index = True,
               left_index = True) # Create a new dataframe by merging average income and population data
income_and_pop_df = income_and_pop_df.reset_index() # Reset the index
```

## Conclusion

In conclusion, there is a correlation between theft frequency and temperature, indoor/outdoor locations, police districts, and time of crime. More extreme temperatures - both hot and cold - tend to discourage theft, along with having less eligible days for theft. Outdoor thefts have prominence over indoor thefts, potentially due to automobile related thefts dominating the dataset. The most dangerous districts for both indoor and outdoor theft were Northeastern and Southeastern, with the safest districts being Eastern and Western. The temperature variation during thefts across districts was negligible due to the insignificant differences in Farenheit. Theft frequencies started to rise and peak during night hours and sunset, whereas they started to bottom out during early hours of the day accompanied by sunrise.

The addition of four more dependent variables allows us to gage Baltimore district safety from various perspectives and metrics. At the police district level, we can see just how much variance there really is among theft rates, share of thefts, theft arrest rates, and crime arrest rates. To reiterate, the differences in these rates could be caused by several variables. Police district funding, staffing, surveillance quality, and work hours are all potential reasons that could contribute to these differing rates. By focusing on the geographic level of police district, we are able to identify areas of weakness that require attention. From here, we can diagnose the problems and test their causality through regression later on.

By integrating average annual household income, total population, and gun offender data, we were able to examine the fluctuation of crime metrics from a different perspective. Utilizing total population to yield per capita measures allowed us to eliminate the bias of population differences across districts. Surprisingly enough, we find that our hypothesis is disproved by graphical data showing the positive correlation between thefts/crimes per capita and average annual household income. These findings may seem irrational, but could ultimately be caused by other endogenous variables that are not accounted for. Following, our hypothesis correctly infers of the negative correlation between gun offenders per capita and average annual household income. It also rightfully deduces the negative correlation between gun offenders per capita and thefts per capita, along with the positive correlation between gun offenders per capita and crimes per capita. In hindsight, we can see that not all endogenous variables and biases are controlled for, leaving some margin of error within our model.

Although many of the conclusions formed seem intuitive, it is important to be cautious of the possibility of endogeneity bias and ommitted variable bias. For instance, the low theft frequencies during hot and cold temperatures could simply be due to only a few days of each year reaching those peak and bottom temperature levels, but this rationale is not easily captured by the graph. Another example is that certain districts may have unobserved characteristics that affect their crime metrics, such as education. A final example of this type of bias is that high theft and crime per capita districts may have higher average annual incomes because of money laundering activity through the black market. Our model fails to account for a plethora of these variables, but it serves as a strong indicator of crime metrics regardless. Without the use of regression, instrumental variables, or any empirical analysis tools at this time, we can only conclude correlation and not causation.
